```python
"""
Homework 3. Statistics
Ryan Quirk <quirkrf@clarkson.edu>
DS 241

1. Calculations
"""
from statistics import *
import numpy as np

dataset = [5,7,8,5,10,12,7,6]
avg = mean(dataset)
med = median(dataset)
mn = min(dataset)
mx = max(dataset)
var = variance(dataset)
dev = stdev(dataset)

print("Dataset:", dataset)
print(f"Mean: {avg}\nMedian: {med}\nRange: {mn}, {mx}\nSample Variance: {var}\nStandard Deviation: {dev}\n")

q1 = np.percentile(dataset, 25)
q3 = np.percentile(dataset, 75)
print(f"25th Percentile: {q1}\n75th Percentile: {q3}\n")

md = multimode(dataset)
print(f"Modes: {md}")
```

## 2. Correlation

```python
from matplotlib import pyplot as plt

X1 = np.array([1,2,3,4,5])
Y1 = np.array([2,4,6,8,10])
coef = np.corrcoef(X1, Y1)
print(f"X1: {X1} ~ Y1: {Y1}")
print("Correlation Coeffecient:", coef[0, 1])
plt.scatter(X1, Y1)
plt.plot(X1,Y1, 'r--')
plt.title(f"X1:Y2, Coef: {round(coef[0, 1])}")
plt.show()


X2 = np.array([1,2,3,4,5])
Y2 = np.array([7,1,4,2,5])
coef = np.corrcoef(X2, Y2)
print(f"X2: {X2} ~ Y2: {Y2}")
print("Correlation Coeffecient:", coef[0, 1])
plt.scatter(X2, Y2)
plt.plot(X2,Y2, 'r--')
plt.title(f"X2:Y2, Coef: {round(coef[0, 1])}")
plt.show()


X3 = np.array([-2,-1,0,1,2])
Y3 = np.array([2,1,0,1,2])
coef = np.corrcoef(X3, Y3)
print(f"X3: {X3} ~ Y3: {Y3}")
print("Correlation Coeffecient:", coef[0, 1])
plt.scatter(X3, Y3)
plt.plot(X3,Y3, 'r--')
plt.title(f"X3:Y3, Coef: {round(coef[0, 1])}")
plt.show()


# Solution doesn't exist
X4 = np.array([])
Y4 = np.array([])
print(np.corrcoef(X4,Y4)[0, 1])
print("No solution exists for empty lists!")


```

## 3. Correlation and Causation

1. The dataset I chose is the Hours Studied vs. Exam Scores

2. Find the statistical correlation between the two variables using python

```python
import csv

rows = []
with open("test-score.csv", 'r') as file:
    reader = csv.reader(file)
    fields = next(reader)

    for row in reader:
        rows.append(row)

hours = []
grades = []

for row in rows:
    hours.append(int(row[0]))
    grades.append(int(row[1]))

hours = np.array(hours)
grades = np.array(grades)

print("Correlation Coeffecient:", np.corrcoef(hours, grades)[0,1])
```

A very strong correlation is found with a coeffecient of 0.986

3. You can find a visualization of the data below.

```python
plt.scatter(hours, grades)
plt.plot(hours,grades, 'r--')
plt.title("Exam Grade for Hours Studied")
plt.show()
```

4. With the correlation coeffecient being so close to 1 at 0.986 this suggests that the number of hours studied before an exam is strongly correlated with the score one gets on their exam.


5. Although there is a very strong correlation between the two variables, this does not imply direct causation between the two variables. For instance, there is a chance that kids who study a lot have better home lives than kids who don't, and maybe this hidden factor has more of an impact on exam grades than hours studied does. There are several factors that go into a variable such as high test grades, and especially low test grades, therefore one can't assume correlation implies causation.

6. The key findings of this analysis is that there is a very strong correlation between hours studied for an exam and the scores achieved on such exam. This is a clear implication of some relationship between variables, but does not imply causation.
